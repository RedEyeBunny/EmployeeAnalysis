```python
#data transformation 
l1 = data['Overall Engagement'].drop_duplicates().tolist()
l2 = data['Job Satisfaction'].drop_duplicates().tolist()

overall_engagement_transformed = data['Overall Engagement'].map({l1[0]:1, l1[1]:2,l1[2]:3,l1[2]:3,l1[3]:4,l1[4]:5})
job_satisfaction_transformed = data['Job Satisfaction'].map({l2[0]:1, l2[1]:2,l2[2]:3,l2[2]:3,l2[3]:4,l2[4]:5})
```

```python
#mean for overall engagement and job satisfaction
# average job satisfaction and overall engagement among employees
print (overall_engagement_transformed.mean())
print (job_satisfaction_transformed.mean())
```

```python
#mode for overall engagement and job satisfaction
# what do most employees feel 
print(overall_engagement_transformed.mode())
print(job_satisfaction_transformed.mode())
```

```python
#standard deviation for overall engagement and job satisfaction
# most employees report similar experiences
print(overall_engagement_transformed.std())
print(job_satisfaction_transformed.std())
```

```python
#likert scale transformation for relevant columns
l1 = data['Overall Engagement'].drop_duplicates().tolist()
l2 = data['Work-Life Balance'].drop_duplicates().tolist()
l3 = data['Compensation Satisfaction'].drop_duplicates().tolist()
l4 = data['Job Satisfaction'].drop_duplicates().tolist()
likert_5_point = {
    "Strongly Disagree":1,
    "Disagree":2,
    "Neutral":3,
    "Agree":4,
    "Strongly Agree":5
}

data['Overall Engagement'] = data['Overall Engagement'].map(likert_5_point)
data['Work-Life Balance'] = data['Work-Life Balance'].map(likert_5_point)
data['Compensation Satisfaction'] = data['Compensation Satisfaction'].map(likert_5_point)
data['Job Satisfaction'] = data['Job Satisfaction'].map(likert_5_point)
data
```

# inferential statistics

```python
from scipy.stats import ttest_ind
from scipy.stats import shapiro

data
```

```python
# Filter the dataset for the specific departments
department_it = data[data['Department'] == 'IT']['Job Satisfaction']
department_hr = data[data['Department'] == 'HR']['Job Satisfaction']

# Test for normality in each department's Job Satisfaction data
from scipy.stats import shapiro

stat_it, p_it = shapiro(department_it)
stat_hr, p_hr = shapiro(department_hr)

print(f"IT Department Normality p-value: {p_it}")
print(f"HR Department Normality p-value: {p_hr}")

# Conduct t-test if normal, otherwise consider Mann-Whitney U test
if p_it > 0.05 and p_hr > 0.05:
    from scipy.stats import ttest_ind
    t_stat, p_value = ttest_ind(department_it, department_hr)
    print(f"T-test p-value: {p_value}")
else:
    from scipy.stats import mannwhitneyu
    u_stat, p_value = mannwhitneyu(department_it, department_hr)
    print(f"Mann-Whitney U Test p-value: {p_value}")
```

```python
# Perform Pearson correlation (or Spearman if data is non-normal)
from scipy.stats import pearsonr

work_life_balance = data['Work-Life Balance']
overall_engagement = data['Overall Engagement']

# Pearson correlation
corr_coef, p_value = pearsonr(work_life_balance, overall_engagement)
print(f"Pearson Correlation Coefficient: {corr_coef}, P-value: {p_value}")
```
